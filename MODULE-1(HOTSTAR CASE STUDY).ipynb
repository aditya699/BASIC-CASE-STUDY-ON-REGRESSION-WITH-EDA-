```python
#REGRESSION MODEL TO PREDICT VIEWS
```

```python
data.columns
```

```python
X=pd.DataFrame(data=data,columns=['Visitors','Views_platform','Ad_impression','Cricket_match_india','Character_A'])
```

```python
y=pd.DataFrame(data=data,columns=['Views_show'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pred))
```

```python
###basic model gives an accuracy of 61%
```
